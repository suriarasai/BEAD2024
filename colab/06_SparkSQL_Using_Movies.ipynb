# SQL Lite

SQLite is a lightweight relational database. SQLite is an in-process library that implements a self-contained, serverless, zero-configuration, transactional SQL database engine. SQLite is a C-language library that provides a lightweight disk-based database, which doesn’t require a separate server process. You can read more about the features of SQLite from https://www.sqlite.org/about.html.

In this tutorial, we will learn to use SQLite on the Google Colab notebook.

To use Python code to create an SQLite database, we start by importing the necessary libraries:

```python
import sqlite3
```

Next, we create a connection to the database (videostore_db). If the database doesn’t exist, SQLite will create it:

```python
conn = sqlite3.connect('videostore_db.sqlite')
```

For this movie tutorial, let’s create a simple table named movies with columns for videocode, title, type, rating, rental price, producer id, director, media, stock, and number rented.

```python
cursor = conn.cursor()
cursor.execute('''
CREATE TABLE IF NOT EXISTS Movies (
    VideoCode    INTEGER      NOT NULL,
    MovieTitle   VARCHAR (50)  NULL,
    MovieType    VARCHAR (10)  NULL,
    Rating       VARCHAR (3)   NULL,
    RentalPrice  DECIMAL (10,2) NULL,
    ProducerID   VARCHAR (50)  NULL,
    Director     VARCHAR (50)  NULL,
    Media        VARCHAR (50)  NULL,
    TotalStock   INTEGER      NULL,
    NumberRented INTEGER      NULL,
    CONSTRAINT PK_Movies PRIMARY KEY(VideoCode)
)
''')
conn.commit()
```
